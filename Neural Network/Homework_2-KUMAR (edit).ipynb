# Homework Assignment #2

The purpose of this assignment is to create two different python functions that implements both the logistic and relu activation functions, make four plots from the four equations, create a function to manipulate the input values, and plot these manipulated input values for a range of slope and y-intercept values for both the logistic and relu activation functions, totaling to another four plots with three graphs on each one. 

##### We are importing the necessary packages as well as defining our input numpy array (integers -8 to 8 with a step value of 0.25)

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from math import e

input = np.arange(-8,8, 0.25)
```

##### Defining a function that takes in an input numpy array, finds the a(n) value for a logistic curve through a series of mathematical functions, and then appends it to a list using a for loop

<font color='blue'>
You should preallocate your arrays for efficiency.
</font>

```python
def logistic_function(input):
    logistic_activation_list = []
    for ind_value in input:
        logistic_activation = 1/(1 + e**(-ind_value))
        logistic_activation_list.append(logistic_activation)
    return logistic_activation_list
```

##### Defining a function that takes in an input numpy array, finds the a(n) value for a logistic curve through vectorization, and assigns it to vect_logistic_output

```python
def vect_logistic_function (input):
    vect_logistic_output = 1/(1+(e**(-input)))
    return vect_logistic_output
```

##### Defining a function that takes in an input numpy array, finds the a(n) value for a relu curve through a series of mathematical functions, and then appends it to a list using a for loop

```python
def relu_function (input):
    relu_activation_list = []
    for ind_value in input:
        if ind_value < 0:
            ind_value = 0
            relu_activation_list.append(ind_value)
        else:
            relu_activation_list.append(ind_value)
    return relu_activation_list
```

##### Defining a function that takes in an input numpy array, finds the a(n) value for a relu curve through vectorization, and assigns it to relu_vect_output

```python
def vect_relu_function (input):
    vect_relu_output = input * (input > 0)
    return vect_relu_output
```

##### Creates a function that takes the parameters--title of graph, label of x axis, label of y axis, font size, color of graph, x-values from the input, and y-values from the operation done on the input--and plots them on a graph

```python
def functions_plot (title, xlabel, ylabel, fontsize, color, inputs, function):
    fig = plt.figure()
    plt.xlabel(xlabel, fontsize = fontsize)
    plt.ylabel(ylabel, fontsize = fontsize)
    plt.title(title, fontsize = fontsize)
    p = plt.plot(inputs, function, color=color) 
```

##### Plots four plots, with each plot corresponding to one of the four cells above. Two of the plots will graph the logistic function using the two different methods--for loop and then vectorization. Two of the other plots will graph the ReLu function using the two methods above as well.

```python
functions_plot ('Logistic Plot with For Loop', 'Input', 'Output',15, 'blue', input, logistic_function(input))
functions_plot ('ReLu Plot with For Loop', 'Input', 'Output', 15, 'green', input, relu_function (input))
functions_plot ('Vectorized Logistic Plot', 'Input', 'Output', 15, 'red', input, vect_logistic_function (input))
functions_plot ('Vectorized ReLu Plot', 'Input', 'Output', 15,'black', input, vect_relu_function (input))
```

##### The below function takes a user's value for slope, an input array, and y-intercept, and causes the input array to undergo a linear transformation from x to mx + b

```python
def input_function (slope, inputs, y_int):
    neuron = (slope*inputs) + y_int
    return neuron
```

##### Creates a function that takes in the parameters for title, a label for the x axis, a label for the y axis, font size, input values (x values), a list of list containing the y values from the operation done to the input values, and the manipulation done to the array, whether that be adjusting the slope or the y-intercept. With this information, a certain number of lines (dependent on the number of manipulations) are plotted on the same figure containing a x label, y label, title, and a figure legend.

```python
def overlap_plot (title, xlabel, ylabel, fontsize, inputs, list_func, manipulation):
    fig = plt.figure()
    plt.xlabel(xlabel, fontsize = fontsize)
    plt.ylabel(ylabel, fontsize = fontsize)
    plt.title(title, fontsize = fontsize)
    graph_list = []
    for ind_func in list_func:
        p, = plt.plot(inputs, ind_func)
        graph_list.append(p)
    plt.figlegend(graph_list, manipulation, bbox_to_anchor=(0.27, 0.88))
```

##### Relu Manipulation of M: a new slope is iterated every time through the vect_relu_function to give three new lines that are plotted using the overlap_plot function

```python
b = 0
slope = [0.5, 1, 1.5]
func_list = []
for i in slope:
    func_list.append(vect_relu_function (input_function (i, input, b)))
overlap_plot ('Relu Manipulation of M', 'Input', 'Ouput', 15, input, func_list, slope)
```

##### Relu Manipulation of B: a new y-intercept is iterated every time through the vect_relu_function to give three new lines that are plotted using the overlap_plot function

```python
slope = 0.5
b = [0, 1, 2]
func_list = []
for i in b:
    func_list.append(vect_relu_function (input_function (slope, input, i)))
overlap_plot ('Relu Manipulation of B', 'Input', 'Ouput', 15, input, func_list, b)
```

```python
b = 0
slope = [0.5, 1, 1.5]
func_list = []
for i in slope:
    func_list.append(vect_logistic_function (input_function (i, input, b)))
overlap_plot ('Logistic Manipulation of M', 'Input', 'Ouput', 15, input, func_list, slope)
```
